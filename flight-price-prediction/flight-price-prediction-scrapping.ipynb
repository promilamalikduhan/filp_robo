```python
from urllib.request import urlopen  
from urllib.request import Request  
from bs4 import BeautifulSoup
import pandas as pd
import time
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
from IPython.display import Image, HTML
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException,StaleElementReferenceException
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.options import Options
import urllib.parse
from selenium.webdriver.support.ui import WebDriverWait
```

```python
Airline=[]
Date_of_Journey=[]
Source=[]
Destination=[]
Stop=[]
Dep_Time=[]
Arrival_Time=[]
Duration=[]
Price=[]

source="DEL"
source_name="Delhi"
destination={"Jaipur":"JAI","Lucknow":"LKO","Srinagar":"SXR",
               "Leh":"IXL"}   
options = Options()
options.add_argument("--headless")
driver = webdriver.Chrome()
dates=['24/07/2022','25/07/2022','26/07/2022','27/07/2022','28/07/2022','29/07/2022','30/07/2022','31/07/2022','01/08/2022',
       '02/08/2022','03/08/2022','04/08/2022','05/08/2022','06/08/2022','07/08/2022','08/08/2022','09/08/2022','10/08/2022',
       '11/08/2022','12/08/2022','13/08/2022','14/08/2022','15/08/2022','16/08/2022','17/08/2022','18/08/2022','19/08/2022',
       '20/08/2022','21/08/2022','22/08/2022',
       '23/08/2022','24/08/2022','25/08/2022','26/08/2022','27/08/2022','28/08/2022','29/08/2022','30/08/2022','31/08/2022',]
ignored_exceptions=(NoSuchElementException,StaleElementReferenceException,)
for key,value in destination.items():
    for m in dates:
        url="https://www.makemytrip.com/flight/search?itinerary="+source+"-"+value+"-"+m+"&tripType=O&paxType=A-1_C-0_I-0&intl=false&cabinClass=E&ccde=IN&lang=eng"
        driver.get(url)
        time.sleep(2)
        try:
            driver.find_element(By.XPATH,'//button[@class="button buttonSecondry buttonBig fontSize12 relative"]').click()
        except NoSuchElementException:
            d=1
        lastHeight = driver.execute_script("return document.body.scrollHeight")
        z=0
        while True:
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            time.sleep(1)
            newHeight = driver.execute_script("return document.body.scrollHeight")
            if newHeight == lastHeight:
                break
            lastHeight = newHeight
            z=z+1
        secs=3
        airline_data1=WebDriverWait(driver, secs).until(EC.presence_of_element_located((By.XPATH,'//p[@class="boldFont blackText airlineName"]')))
        airline_data=driver.find_elements(By.XPATH,'//p[@class="boldFont blackText airlineName"]')
        time_data1=WebDriverWait(driver, secs).until(EC.presence_of_element_located((By.XPATH,'//p[@class="blackText fontSize18 blackFont appendBottom2 makeFlex hrtlCenter"]')))
        time_data=driver.find_elements(By.XPATH,'//p[@class="blackText fontSize18 blackFont appendBottom2 makeFlex hrtlCenter"]')
        duration_data1=WebDriverWait(driver, secs).until(EC.presence_of_element_located((By.XPATH,'//div[@class="stop-info flexOne"]')))
        duration_data=driver.find_elements(By.XPATH,'//div[@class="stop-info flexOne"]')
        stop_data1=WebDriverWait(driver, secs).until(EC.presence_of_element_located((By.XPATH,'//p[@class="flightsLayoverInfo"]')))
        stop_data=driver.find_elements(By.XPATH,'//p[@class="flightsLayoverInfo"]')
        price_data1=WebDriverWait(driver, secs).until(EC.presence_of_element_located((By.XPATH,'//p[@class="blackText fontSize18 blackFont white-space-no-wrap"]')))
        price_data=driver.find_elements(By.XPATH,'//p[@class="blackText fontSize18 blackFont white-space-no-wrap"]')
        for i in range(len(airline_data)):
            Airline.append(airline_data[i].text)
            duration_info = duration_data[0].text.split("\r\n")
            Date_of_Journey.append(m)
            Source.append(source_name)
            Destination.append(key)
            Stop.append(stop_data[i].text)
            Dep_Time.append(time_data[i].text)
            Arrival_Time.append(time_data[i+1].text)
            Duration.append(duration_info[0].split("\n")[0])
            Price.append(price_data[i].text)
            i=i+2
        df=pd.DataFrame({'Airline':Airline,'DOJ':Date_of_Journey,'Source':Source,'Destination':Destination,'Stop':Stop,
                        'Departure':Dep_Time,'Arrival':Arrival_Time,'Duration':Duration,'Price':Price})

        df 
        df.to_csv('flight_data.csv', mode='a', header=False)
```
